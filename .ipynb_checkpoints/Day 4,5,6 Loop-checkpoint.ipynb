```python
s = str()
i = 0
n = int(input("Enter n ="))
while i<n:
    name = input("Enter name = ")
    phone = int(input("Enter phone = "))
    s= s+name+" "+str(phone)+'\n'
    i = i+1
print(s)
```

```python
s = str()
all_total = 0
i = 0
n = int(input("Enter n ="))
while i<n:
    product = input("Enter name = ")
    price = int(input("Enter product price = "))
    quantity = int(input("Enter quantity ="))
    total = price*quantity
    all_total = all_total+total
    s= s+product+" "+str(price)+" "+str(quantity)+" "str(total)+"'\n'
    i = i+1
print(s)
print("All total =",all_total)
```

```python
# Break and continue in while loop
a = "Hello world i am python"
l = len(a)
i = 0
while i<l:
    if a[i] == " ":
        break
    print(a[i],end ="")
    i = i+1
```

```python
# Break and continue in while loop
a = "Hello world i am python"
l = len(a)
i = 0
while i<l:
    if a[i] == " ":
        i = i+1
        continue 
        
    print(a[i],end ="")
    i = i+1
```

```python
# prime = 2,3,5,7,11,13,17
# composite = 4,6,8,10,17 


n = int(input("Enter n = "))
if n>=2:
    for i in range(2,n//2):
        if n%i == 0:
            print(n,"is composite no.")
        break 
    else:
        print(n,"is prime")
    
else:
    print("The value of n should be greater or equal to 2")
```
